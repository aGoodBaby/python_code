# 决策树实战
1、基于iris_data.csv数据，建立决策树模型，评估模型表现  
2、可视化决策树结构  
3、修改min_samples_leaf参数，对比模型结果

```python
#load the data
import pandas as pd
import numpy as np
data = pd.read_csv('iris_data.csv')
data.head()
```

```python
#define X,y
X = data.drop(['target', 'label'], axis=1)
y = data.loc[:, 'label']
#establish the decision tree model
from sklearn import tree
dc_tree = tree.DecisionTreeClassifier(criterion='entropy', min_samples_leaf=5)
dc_tree.fit(X, y)
#evaluate the model
y_predict = dc_tree.predict(X)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, y_predict)
accuracy
```

```python
#visualize the tree
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(10, 10))
text = tree.plot_tree(
    dc_tree,
    filled='True',
    feature_names=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'],
    class_names=['setosa', 'versicolor', 'virginica'])
```

```python
#change min_samples_leaf
dc_tree = tree.DecisionTreeClassifier(criterion='entropy', min_samples_leaf=10)
dc_tree.fit(X, y)
y_predict = dc_tree.predict(X)
accuracy = accuracy_score(y, y_predict)
print(accuracy)
fig = plt.figure(figsize=(10, 10))
text = tree.plot_tree(
    dc_tree,
    filled='True',
    feature_names=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'],
    class_names=['setosa', 'versicolor', 'virginica'])
```
