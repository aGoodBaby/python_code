```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../Data/usa_flights.csv')
df['delayed'] = df['arr_delay'].apply(lambda x:x>0)
df=df[['flight_date','unique_carrier','flight_num',
       'arr_delay','cancelled','delayed']]
df.head()
```

```python
#1.获取原物时间最长top5
df.sort_values('arr_delay',ascending=False)[:5]
```

```python
#2.计算有延误和没有延误的所占比例
delay_data = df['delayed'].value_counts()
delay_data[1]/(delay_data[0]+delay_data[1])
```

```python
#3.每一个航空公司延误的情况
delay_group = df.groupby(['unique_carrier','delayed'])
delay_group.size().unstack().head()#group.size()计算分组的数量
```

```python
#4.透视表功能
flights_by_carrier = df.pivot_table(index='flight_date',
            columns='unique_carrier',values='flight_num',aggfunc='count')
flights_by_carrier.head()
```
