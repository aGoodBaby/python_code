```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Data/apply_demo.csv')
df.head()
```

```python
#初识apply
df['A'] = pd.Series(['a']*df.size)
df['A'] = df.A.apply(str.upper)#df.A好像不能赋值
df.head()
```

```python
#********将data数据拆分*****
def foo(line):
    items = line.strip().split(' ')
    return pd.Series([items[1],items[3],items[5]])
df_tmp = df.data.apply(foo)
df_tmp = df_tmp.rename(columns={0:'Symbol',1:'Seqno',2:'Price'})
df_tmp.head()
```

```python
df_new = pd.concat([df,df_tmp],axis=1)
del df_new['data']
del df_new['A']
df_new.head()
```

```python
df_new.to_csv('../Data/demo_duplicate.csv',index=False)
```
