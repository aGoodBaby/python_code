```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#直方图，与柱状图不同，这里显示的是数据分布
s = pd.Series(np.random.randn(1000))
s.plot(kind='hist',rwidth=0.9)
```

```python
#密度图
s.plot(kind='kde')
```

```python
#散点图
iris = pd.read_csv('../Data/iris.csv')
iris.head()
```

```python
for species,group in iris.groupby('Name'):
    plt.scatter(group['PetalLength'],
                group['SepalLength'],label=species)
plt.legend(title='Name')
plt.xlabel('petalLength')
plt.ylabel('sepalLength')
```
