```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('flights')
df.head()
```

```python
df = df.pivot_table(index='month',columns='year',values='passengers')
df
```

```python
sns.set()
sns.heatmap(df,annot=True,fmt='d')
```

```python
s=df.sum()
sns.barplot(x=s.index,y=s.values)
```
