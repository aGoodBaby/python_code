```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,14,100)
y1= np.sin(x)
y2= np.sin(x+2)*1.25
def sinplot():
    plt.plot(x,y1)
    plt.plot(x,y2)
sinplot()
```

```python
import seaborn as sns
```

# <font color=steelblue>axes_style() and set_style()</font>

```python
style = ['darkgrid','dark','white','whitegrid','ticks']
sns.set_style(style[3],{'grid.color':'blue'})
sinplot()
```

```python
#这些属性都可以通过上面这种方式修改
sns.axes_style()
```

```python
#重置为默认显示效果
sns.set()
sinplot()
```

# <font color=steelblue>plotting_context() and set_context()</font>

```python
context = ['paper','notebook','talk','poster']#设置显示大小
sns.set_context(context=context[3],rc={'grid.linewidth':5.0})
sinplot()
```
