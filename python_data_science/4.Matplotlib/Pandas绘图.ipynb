```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
s1 = pd.Series(np.random.randn(10)).cumsum()
s2 = pd.Series(np.random.randn(10)).cumsum()
```

```python
s1.plot(kind='line',grid=True,label='s1',title='this is series')
s2.plot(label='s2',kind='line',grid=True,legend=True)
```

```python
fig,ax = plt.subplots(1,2)
s1.plot(ax=ax[0],kind='bar',label='s1')
s2.plot(ax=ax[1],label='s2')
fig.legend()
```
