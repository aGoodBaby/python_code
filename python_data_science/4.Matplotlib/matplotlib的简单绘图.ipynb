```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,20)
y1= np.sin(x)
y2= np.sin(x*2)
plt.figure(figsize=(8,6))
plt.plot(x,y1)
```

```python
plt.plot(x,y1,'r--',label='aaa')
plt.plot(x,y2,'b--',label='bbb')
plt.xlabel('this is x');plt.ylabel('this is y')
plt.title('this is a title')
plt.legend()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,'r--',label='aaa')
plt.ylabel('y1')
plt.legend()
plt.subplot(212)
plt.plot(x*2,y2,'b--',label='bbb')
plt.ylabel('y2')
plt.xlabel('x')
plt.legend()
```
