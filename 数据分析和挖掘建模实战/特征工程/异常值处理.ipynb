```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"A": ["a0", "a1", "a1", "a2", "a3", "a4"],
                   "B": ["b0", "b1", "b2", "b2", "b3", None],
                   "C": [1, 2, None, 3, 4, 5], 
                   "D": [0.1, 10.2, 11.4, 8.9, 9.1, 12], 
                   "E": [10, 19, 32, 25, 8, None],
                   "F": ["f0", "f1", "g2", "f3", "f4", "f5"]})
df
```

```python
#判断是否空值
df.isnull()
```

```python
#去除空值
df.dropna(subset=['C'])#都是生成新的df，不修改原表
```

```python
#是否重复值
df.duplicated(['A'])
```

```python
#去除重复值
df.drop_duplicates(['A'],keep='first')
```

```python
#填充空值
df['B'].fillna('b*')
```

```python
df['E'].fillna(df['E'].mean())
```

```python
#插值
df['E'].interpolate(method='linear',order=3)
```

```python
#超出边界的异常值
q_inter = df['D'].quantile(0.75)-df['D'].quantile(0.25)
df[df['D']<df['D'].quantile(0.75)+1.5*q_inter]\
  [df['D']>df['D'].quantile(0.25)-1.5*q_inter]
```

```python
#知识异常值
df[[True if item.startswith('f') else False for item in list(df['F'].values)]]
```
